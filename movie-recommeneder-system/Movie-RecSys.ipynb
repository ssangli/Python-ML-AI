 # Recalculate vote_average for the reduced dataset
 Formula : 
 v = vote_count for the movie
 R = vote_average for the movie
 m = 95th percentile vote_count value
 C = mean of vote_average
 
 new_rating = (v / (m + v) ) * R + (m / (m + v)) * C

```python
mmd_df2 = mmd_df1[['title','vote_count', 'vote_average', 'id', 'imdb_id', 'popularity', 'overview', 'genres', 'release_date']]
```

```python
mmd_df2.shape
```

```python
m = vote_counts.quantile(0.95)
C = movie_metadata['vote_average'].mean()
```

```python
m, C
```

```python
def weighted_voting_rate(x):
    v = x['vote_count']
    R = x['vote_average']
    return ((v/(v+m)) * R + (m/(v+m)) * C)
```

```python
mmd_df2['weighted_rating'] = mmd_df2.apply(weighted_voting_rate, axis=1)
```

```python
mmd_df2.shape
```

```python
mmd_df2.head()
```

```python
# Rating for movie DDLJ
ddlj_rating = mmd_df2[mmd_df2['vote_average'] == 9.1][['title', 'vote_count', 'vote_average', 'weighted_rating']]
inception_rating = mmd_df2[mmd_df2['vote_count'] == 14075.0][['title', 'vote_count', 'vote_average', 'weighted_rating']]
```

```python
print("DDLJ {}\n Inception {}".format(ddlj_rating, inception_rating))
```

After scaling, ddlj(7.72) rating is lower than that of inception (8.02)
